```python
import pandas as pd

with open("data-faces-wider/annotations.txt") as f:
    lines = []
    for line in f:
        lines.append(line.strip('\n'))

df = pd.DataFrame(columns=['x', 'y', 'w', 'h', 'f', 'r', 'e', 't', 'p','name'])
img='ii'
for line in lines:
    if 'jpg' in line:
        img = line
    elif ',' in line:
        position = line.split(',')
        position.append(img)
        df.loc[len(df),:]=position
```

```python
adf=(df[df['name']=='0.jpg'])
```
